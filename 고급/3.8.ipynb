```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation as plt2
from matplotlib import rc
```

```python
g=np.pi**2
v0=10.
dt=0.01
theta=np.pi/3
vx0=v0*np.cos(theta)
vy0=v0*np.sin(theta)

ke0=.5*(v0**2)
ke=ke0
b=0

while (ke>5*ke0/6):
    x=0.;y=0.
    vx=vx0;vy=vy0
    px1=[];py1=[]
    while y>-1e-5:
        x+=vx*dt
        y+=vy*dt
        
        vx+= -b*vx*dt
        vy+= -b*vy*dt -g*dt
        
        px1.append(x)

        py1.append(y)
    ke=.5*(vx**2+vy**2)
    b+=0.0001
print(b, ke/ke0)
```
