```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import rc
rc('animation', html='jshtml')
%matplotlib inline 
```

```python
g=np.pi**2
b=0.1
v0=10.

theta=np.pi/2
vx0=v0*np.cos(theta)
vy0=v0*np.sin(theta)
x=0.;y=0.
vx=vx0;vy=vy0
print(vx,vy)
ke0=.5*(v0**2) * 2   # 초기 운동에너지 입자가 2개이기에 곱하기2


# 수직으로 쏴서 터지기전에 올라가는 좌표를 고려하여 리스트에 넣기

g=np.pi**2
b=0.1
v0=20.

theta=np.pi/2
vx0=v0*np.cos(theta)
vy0=v0*np.sin(theta)
x=0.;y=0.
vx=vx0;vy=vy0

px1=[]; py1=[]
px2=[]; py2=[]
px3=[]; py3=[]
px4=[]; py4=[]

dt=0.001


z=0
while vy>0:
        x+=0 #vx*dt
        y+=vy*dt
        
        vx+= -b*vx*dt
        vy+= -b*vy*dt -g*dt
        px2.append(x)
        py2.append(y)
        px1.append(x)
        py1.append(y)
        px3.append(x)
        py3.append(y)
        px4.append(x)
        py4.append(y)
        z+=1

        
ke=.5*(vx**2+vy**2)*2   #10 위에서의 운동 에너지,  ㅇ비자가 2개니가 *2
print(ke)

ke=ke+(ke0)*2
print(ke)

v02=np.sqrt(ke)

theta1=np.pi*np.random.rand(1)
theta2=theta1+np.pi/2
theta3=theta2+np.pi/2
theta4=theta3+np.pi/2


dt=0.001
vx10=v02*np.cos(theta1)
vy10=v02*np.sin(theta1)
vx20=v02*np.cos(theta2)
vy20=v02*np.sin(theta2)
vx30=v02*np.cos(theta3)
vy30=v02*np.sin(theta3)
vx40=v02*np.cos(theta4)
vy40=v02*np.sin(theta4)

vx1=vx10;vy1=vy10
vx2=vx20;vy2=vy20
vx3=vx30;vy3=vy30
vx4=vx40;vy4=vy40

x1,y1=x,y
x2,y2=x,y
x3,y3=x,y
x4,y4=x,y




z=0
while ((y1)+(y2)+(y3)+(y4))>-3e2:
        x1+=vx1*dt; y1+=vy1*dt
        x2+=vx2*dt; y2+=vy2*dt
        x3+=vx3*dt; y3+=vy3*dt
        x4+=vx4*dt; y4+=vy4*dt
        
        vx1+= -b*vx1*dt; vy1+= -b*vy1*dt -g*dt
        vx2+= -b*vx2*dt; vy2+= -b*vy2*dt -g*dt
        vx3+= -b*vx3*dt; vy3+= -b*vy3*dt -g*dt
        vx4+= -b*vx4*dt; vy4+= -b*vy4*dt -g*dt
        
        px1.append(float(x1))
        py1.append(float(y1))
        px2.append(float(x2))
        py2.append(float(y2))
        px3.append(float(x3))
        py3.append(float(y3))
        px4.append(float(x4))
        py4.append(float(y4))
        # print(x1,x2,y1,y2)
        
# plt.axis([-1,1, 17,19])

plt.plot(px1,py1,
         px2,py2,
         px3,py3,
         px4,py4
        )


```
