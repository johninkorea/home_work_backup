은하의 경우 자외선 초과

```python
# import modules.
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
# Define constants.
h=6.6260755*(1e-27)  #Planck constant(erg s)
c=2.99793458*(1e18)  #Speed of light(A/s)
k=1.380658*(1e-16)   #Boltzmaann constant(erg/K)
e=np.e               #Euler's constant
R0=6.95508*(1e18)  #Solar radius(A)
num=1e9              #Number of stars
T=[6e3, 3e4]        #Temperature of sun and BHB(K)
```

```python
s=np.pi*((R0)**2)      #square measure of sun
s1=np.pi*((50*R0)**2)  #square measure of BHB
```

```python
def B_lamda(lamda,Temperature,square_measure,number_of_star):#Define (function planck distribution)*(square_measure*number).
    B=(2*h*(c**2))*(lamda**-5)/(e**((h*c)/(lamda*k*Temperature))-1)*(square_measure*number_of_star)/(4*np.pi)
    return B
```

```python
lamda=(np.arange(1e2,8e3))  #Define range of lamda.

# Define list that be append result of function.
y=[]      
y1=[]
y10=[]
y100=[]
y1000=[]

# make results whith while.
z=0
while z<len(lamda):
    a=log10(B_lamda(lamda[z], T[0],s,num))
    a1=log10(B_lamda(lamda[z], T[0],s,num)+B_lamda(lamda[z], T[1],s1,1))
    a10=log10(B_lamda(lamda[z], T[0],s,num)+B_lamda(lamda[z], T[1],s1,10))
    a100=log10(B_lamda(lamda[z], T[0],s,num)+B_lamda(lamda[z], T[1],s1,100))
    a1000=log10(B_lamda(lamda[z], T[0],s,num)+B_lamda(lamda[z], T[1],s1,1000))
    
    y.append(a)
    y1.append(a1)
    y10.append(a10)
    y100.append(a100)
    y1000.append(a1000)
    z+=1
```

```python
# plot results.
plt.axis([0,8000, 30,40])
plt.title('The UV-upturn')
plt.xlabel('λ(Å)')
plt.ylabel('log Bλ(erg/Å/s/sr)')
plt.plot(lamda,y,label='10^9 Sun')
plt.plot(lamda,y1,label='10^9 Sun + 1BHB')
plt.plot(lamda,y10,label='10^9 Sun + 10BHB')
plt.plot(lamda,y100,label='10^9 Sun + 100BHB')
plt.plot(lamda,y1000,label='10^9 Sun + 1000BHB')
plt.legend(loc='best')
```
